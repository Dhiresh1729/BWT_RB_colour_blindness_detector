# GENE FREQUENCY DISTRIBUTION 

to print the number lines, genes, composition of each gene & plot a pie-chart of the gene composition for a given file

```python
import matplotlib.pyplot as plt
import math
```

```python
def print_char_freq(file_name):
    # print the genomic composition in a form of a pie-chart for a given file

    # Parameters:
    # file_name: name of the file

    file = open('./' + str(file_name), 'r') #opening the file

    freq = {}
    num_lines = 0
    num_chars = 0
    genes = {'A', 'C', 'G', 'T'}

    while 1:
        char = file.read(1) #reading file char by char
        if not char:
            break

        if char in genes:

            if char in freq:
                freq[char] += 1
            else:
                freq[char] = 1

            num_chars += 1

        elif char == '\n':
            num_lines += 1


    file.close()

    total_num_gene = 0

    for char in freq:
            total_num_gene += freq[char]

    print(f"for file: {file_name}") 
    print(f"total number of lines: {num_lines/1E6}M")
    print(f"total number of characters: {num_chars/1E6}M\n")

    for char in freq:
            print(f"number of {char}: {freq[char]/1E6}M ({round(freq[char]/total_num_gene*100,2)}%)")

    fig = plt.figure(figsize = (8,5)) #plotting the pie-chart for the genomic composition
    plt.title(f"Gene frequency distribution in {file_name}")
    plt.pie(freq.values(),
            labels=freq.keys(),
            autopct='%.2f%%')
```

```python
print_char_freq("chrX.fa")
```

```python
print_char_freq("reads")
```

# READS LENGTH V/S COUNT BAR GRAPH

a bar graph representing length vs log10(count) of reads in the reads file

```python
def print_reads_histogram():
    # print the bar graph represeting legnth vs log10(count) 
    # of reads in the reads file

    file = open('reads', 'r') #opening the reads file
    lines = file.readlines()
    file.close()

    l = 0
    freq = {}

    for line in lines:

        if len(line) in freq:
            freq[len(line)] += 1
        else:
            freq[len(line)] = 1

        l += len(line)

    values = [math.log10(value) for key, value in freq.items()] 
    fig = plt.figure(figsize=(12,8))

    plt.bar(freq.keys(), values, width = 0.6) #plotting the bar graph for reads length v/s log10(count)
    plt.xlabel("Length of reads")
    plt.ylabel("Log10(count) of reads")
    plt.title("reads length v/s log10(count)")
    plt.show()

    print(f"Average read length: {l/len(lines)}")
```

```python
print_reads_histogram()
```
